### Alphabet Pattern Size Input

In this section, we take user input for the size of the alphabet pattern.

- The user is prompted to enter a size.
- If the size is even, it is incremented by 1 to ensure symmetry in the pattern.
- We create two lists:
  - `y`: A list of numbers from 0 to the specified size.
  - `x`: A reversed version of `y`, which will be used in generating the pattern.


```python
size=int(input("Enter a size of alphabets: "))
if size%2==0:
    size=size+1
y=[i for i in range(0,size)]
x=y[::-1]
```

### Function to Print Alphabet 'A' Pattern

The `A()` function generates and prints a pattern that resembles the letter 'A'.

- **Pattern logic:**
  - The top row (`i == 0`) is completely filled with `*` symbols.
  - The middle row (`i == len(y) // 2`) is also fully filled, representing the horizontal bar of the letter 'A'.
  - The leftmost (`j == 0`) and rightmost (`j == len(y) - 1`) columns are filled with `*`, creating the sides of the 'A'.
  - All other positions are filled with spaces to shape the internal structure of the pattern.


```python
def A():
    for i in y:
        for j in y:
            if i==0 or i==len(y)//2 or j==0 or j==len(y)-1:
                print("*",end=" ")
            else:
                print(" ",end=" ")
        print()
```

### Alphabet 'B' Pattern

This function prints the letter 'B' using the following logic:

- Top, middle, and bottom rows are filled with `*` (except the last column).
- Leftmost column is always `*`.
- Rightmost column is `*` except in the top, middle, and bottom rows.


```python
def B():
    for i in y:
        for j in y:
            if ((i==0 or i==len(y)//2 or i==len(y)-1) and j<len(y)-1) or j==0 or (j==len(y)-1 and not(i==0 or  i==len(y)//2 or i==len(y)-1)) :
                print("*",end=" ")
            else:
                print(" ",end=" ")
        print()
```

### User Input and Alphabet Pattern Display

This section takes user input for a name and prints the corresponding alphabet patterns.

- The input name is converted to uppercase.
- For each character in the name, a match statement determines which alphabet function to call:
  - Each letter from 'A' to 'Z' calls its respective function (e.g., `A()`, `B()`, etc.).
- If the character is not recognized, it prints a blank space.

```python
a= input ("Enter a name to print: ")
a=a.upper()
for i in a :
    print("\n")
    match i:
        case 'A':
             A()
        case 'B':
             B()
        case 'C':
             C()
        case 'C':
             C()
        case 'D':
             D() 
        case 'E':
             E()
        case 'F':
             F()
        case 'G':
             G()
        case 'H':
             H()
        case 'I':
             I()
        case 'J':
             J()
        case 'K':
             K()
        case 'L':
             L()
        case 'M':
             M()
        case 'N':
             N()
        case 'O':
             O()
        case 'P':
             P()
        case 'Q':
             Q()
        case 'R':
            R()
        case 'S':
             S()
        case 'T':
             T()
        case 'U':
             U()
        case 'V':
             V()
        case 'W':
             W()
        case 'X':
             X()
        case 'Y':
             Y()
        case 'Z':
             Z()
        
        case _:
            print("",end=' ')
```
